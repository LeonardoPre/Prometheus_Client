```python
import pandas as pd

def get_interval_duration(df, column_name="observation_time"):
    """
    Calculates the total duration of a workload in seconds.
    """
    # 1. Convert the column to datetime objects
    times = pd.to_datetime(df[column_name])
    
    # 2. Calculate the difference between the latest and earliest timestamp
    duration = times.max() - times.min()
    
    # 3. Return the total seconds as a float
    return duration.total_seconds()

duration_shaped = get_interval_duration(df_pods_shaped)
duration_fixed = get_interval_duration(df_pod_fixed)

print(f"Shaped workload duration: {duration_shaped}s")
print(f"Fixed workload duration: {duration_fixed}s")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Access the Viridis colormap
# We'll take two colors: one at 20% and one at 80% of the scale
cmap = plt.get_cmap('viridis')
color_v09 = cmap(0.2)   # Deep Purple/Blue
color_v11 = cmap(0.8)   # Bright Yellow/Green

data_to_plot = [
    df_pods_shaped["wattage_kepler"], df_pods_shaped["wattage_kepler_new"],
    df_pods_pausing["wattage_kepler"], df_pods_pausing["wattage_kepler_new"],
    df_pod_rampup["wattage_kepler"], df_pod_rampup["wattage_kepler_new"],
    df_pod_fixed["wattage_kepler"], df_pod_fixed["wattage_kepler_new"]
]

positions = [1, 2, 4, 5, 7, 8, 10, 11]
fig, ax = plt.subplots(figsize=(12, 6))

# 2. Create the boxplot
bp = ax.boxplot(data_to_plot, positions=positions, widths=0.6, patch_artist=True,
                medianprops={"color": "black"})

# 3. Apply Viridis colors alternately
colors = [color_v09, color_v11] * 4
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.8) # Slight transparency looks cleaner

# 4. Styling
ax.set_ylabel('Wattage')
ax.set_xticks([1.5, 4.5, 7.5, 10.5])
ax.set_xticklabels(['Shaped', 'Pausing', 'Rampup', 'Fixed'])

# 5. Legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor=color_v09, label='Kepler 0.9.0'),
    Patch(facecolor=color_v11, label='Kepler 0.11.2')
]
ax.legend(handles=legend_elements, loc='upper right')

plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()

# Save as .eps for that high-quality vector format we discussed!
fig.savefig('figs/kepler_comparison_viridis.eps', format='eps', bbox_inches='tight')

plt.show()
```
